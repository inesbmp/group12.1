### Introduction

SAIL Amsterdam is a large 5-day maritime festival held every five years. The event attracts over 10,000 ships and more than 2.3 million visitors. Managing crowds of this size requires continuous crowd management to prevent issues such as congestion, bottlenecks and overcrowding at specific locations.

There already are a number of sensors located at the event to support crowd managers. However, a user-friendly platform such as a dashboard, has not been made yet.
This project focuses on developing a dashboard that helps crowd managers monitoring crowd flows and predicting potential issues for taking action on time.

This proposal outlines the project’s objectives, defines the project’s scope, formulates the research questions and describes the intended project pipeline.


### Objectives

The objectives of this project were derived from the SAIL lecture, existing documentation, and discussions with the TAs. Based on insights from these sources, three primary objectives were defined; each has its own purpose and relevance to the project. The research question and its sub‑questions shown later in this report were developed with these objectives in mind.

The first objective is the creation of an intuitive visualisation dashboard with relevant real time data. It is essential and the ultimate goal of this project to have a dashboard for the crowd managers to see, from which they must be able to analyse the situation and make decisions according to the given information, for example by the means of a graph, map and warnings. This visualisation has to be intuitive, giving a clear oversight of the situation and easy to interpret. Irrelevant information should not be shown and can possibly shift the focus from the crowd managers away from more important information. Therefore only relevant data should be processed and shown in the dashboard. At last, for the dashboard to be useful for the crowd supervisors during an event, the information shown in the dashboard should be from real time data and must have the ability to be updated with newly collected data. 

The second objective is the writing of code and the development of algorithms to process real time data into relevant information. The transition from the gathered raw data into the graphs, maps and warnings requires programming. Multiple algorithms will need to be developed in order for this data to be processed into relevant and useful information and this will involve writing code in a programming language. These algorithms should be able to process and adapt to real time data, and thus being able to show relevant information at a certain timepoint for the crowd managers.

The third and last objective is the gathering of relevant data from the Sail project. As this information is, according to the TA, the property of the municipality of Amsterdam, this information has to be requested and approved to be used for this project. Relevant information should be identified and collected in order to be processed by the algorithms.


### Geographical scale


The SAIL event spans across the IJ river as it goes through Amsterdam. Specifically where the river flows by the main station and its surroundings. Neighbouring areas are used to host related events to SAIL. Thus, SAIL’s interest areas are divided as seen in the figure below. 


The blue area (i.e. ‘Hospitality’) is dedicated to sponsor lounges, business hosting, and premium guest activities. The green area (i.e. ‘Navigating the Future ') aims to showcase maritime technology, host exhibits and talks, future shipping, sustainability, etc. The white area (i.e. ‘Young and Urban) hosts a younger-audience program with music, interactive media, urban culture, and public engagement art. The orange area (i.e. ‘The Heart’) is the core of the event, visitors can walk along the docked ships, board vessels, and engage with ship exhibitions. Lastly, the red area (i.e. ‘The Living Room’) has a more relaxed, small scale, and laid-back program with lounges, installations and quieter performances. 

Most of the crowds are located in the orange area. Through that part of the river, is where people gather to watch the ships sailing through. Since this area is the main crowd attractor, most of the sensors dedicated at crowd control are hence positioned around the area. Therefore, the focus of this project will be subject to the ‘Heart’ area. 

When it comes to the temporal scale of the project, it will cover the duration of the event. Namely, 20th - 24th August 2025. 


### Research questions

The main research question and the sub questions were formulated based on the objectives that were described earlier. The main objective is to inform the crowd managers of SAIL about the number of people at certain locations and when these locations start to become too crowded or are expected to be in the future. Therefore the main research question is: How to efficiently inform crowd managers about the current and future busyness of SAIL 2025 via an informative dashboard?

To answer this research question, three sub questions were formulated to help address this question. The sub questions are described as follows: 

1. What data and techniques should be used for data processing?  
2. How should the processed data be handled and what kinds ofalgorithms should be implemented?  
3. What are the best visualization methods to efficiently inform authorities?  

The first sub question concerns what and how the raw data should be processed to convert it to useful information. The second sub question covers how the converted raw data should be used in the dashboard to inform the crowd managers. The third sub question describes how the data should be visualized in the dashboard. Various visualization methods exist, but these should be chosen with careful consideration.


### Intended project pipeline 

The pipeline towards the proposed objectives starts with analyzing the data provided (and the format to be provided in real time) and designing a Data Pipeline suitable to treat the raw, live data into usable information. 

The second part of the project pipeline is to design an intuitive Data Visualization interface (Dashboard), with functionalities such as geographical Crowd-headmap and time projections, as well as attractivity measures for the different parts of the SAIL event. Functionalities will be designed according to the data available.

The last part of the project consists of implementing the designed Data Pipeline and Visualization Interface, as well as the algorithms needed to transform the treated Data (such as pedestrian sensor counts) into usable information (such as amount of pedestrians in each node).
